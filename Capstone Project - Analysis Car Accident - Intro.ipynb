# **Introduction :**

* From this project, we want to analyze existing collision data provided by SPD, 
* to understand what are the most common causes and their correlation; 
* to predict the different accidents' severity in future.

# **Data :**

* This collisions data is been provided by Seattle Police Department; and weekly recorded by Traffic Records from 2004 to present.
* It contains information such as severity code, address type, location, collision type, weather, road condition, speeding, and others.
* The target audiences of this study are those people who really care about the traffic records, especially in the transportation department. 
* Also, we want to analysize the reason for collisions and help to reduce accidents in the future.
